```python
import pandas as pd

df = pd.read_csv('https://public.tableau.com/app/sample-data/enigma-us.gov_.census.eits_.ressales-e7513e56d76050c05caf638306055c98_1.csv')
df.info()
#20404 records, 20 columns, a few columns with null values
#this is a very interesting dataaset, per_name is month, start with cat_code: sold, for sales and annual rate. then it is further broken down to different categories (dt_code), total seems to be what I need
df.head(40)
df[df['geo_code'] == 'US']
df[(df['geo_code'] == 'US') & (df['dt_desc'].notnull())].tail(40)
df[(df['geo_code'] == 'US') & (df['dt_desc'].notnull())].head(40) #avg price doesn't start in 1963, median does
df[(df['geo_code']== 'US') & (df['dt_desc'].notnull()) & (df['dt_code'] == 'MEDIAN')].head(40)
df[(df['geo_code']== 'US') & (df['dt_desc'].notnull()) & (df['dt_code'] == 'AVERAG')].tail(40)
```

```python
df1 = df[(df['geo_code']== 'US') & (df['dt_desc'].notnull()) & (df['dt_code'] == 'AVERAG')]
df1.drop(columns=['et_code','et_desc','et_unit'],inplace=True)
df1.info()
df1.head(10)


#try to get the sales volume
df[(df['geo_code'] == 'US') & (df['cat_code'] == 'SOLD') & (df['dt_code'] == 'TOTAL')].tail(40)
df[(df['geo_code'] == 'US') & (df['dt_code'].notnull())].head(40)
df[(df['geo_code'] == 'US')].head(40)
#df[(df['geo_code'] == 'US') & (df['dt_code'].notnull()) & (df['cat_code'] == 'SOLD') & (df['dt_code'] == 'TOTAL')].tail(40)
#but this data no matter how i look at it, does not make sense if I look at # of units sold, it has stayed pretty much flat over the decades
```

```python
#transform per_name field to date and isolate year
df1['per_name'] = pd.to_datetime(df1['per_name'])
df1.head(20)
df1.info()

df1['Year'] = df1['per_name'].dt.year
df1.head(20)
df1['Month'] = df1['per_name'].dt.month
df1.tail(30)
#then do pivot
pd.pivot_table(df1,values='val',index=['Year'],aggfunc='mean')
pd.pivot_table(df1,values='val',index=['Month','Year'],aggfunc='mean')
df1.info()
df1.loc[df1['Year'] == 2006]
df1.head(40)


#visualize
import matplotlib.pyplot as plt
plt.plot(df1['per_name'],df1['val'],color='red')
plt.xlabel('Dates')
plt.ylabel('Avg Selling Price')
plt.title('Avg Selling Price from 1975 to 2006')
plt.show()


df2 = df[(df['geo_code']== 'US') & (df['dt_desc'].notnull()) & (df['dt_code'] == 'MEDIAN')]
df2['per_name'] = pd.to_datetime(df2['per_name'])
df2 = df2.loc[df2['per_name'] >= '1975-01-01']
```

```python
#plot 2 slide by slide with filler in between

plt.plot(df1['per_name'],df1['val'],color='red')
plt.plot(df2['per_name'],df2['val'],color='blue')
plt.xlabel('Dates')
plt.ylabel('Avg Selling Price')
plt.title('Avg Selling Price from 1975 to 2006')
plt.fill_between(df1['per_name'],df1['val'],df2['val'],color='green', alpha = 0.2)
plt.show()
#the gap between median price and avg price is increasing over time, meaning there are more luxury high end properties being sold
#housing crisis in 2008 is also apparent in the chart, pretty cool
```
